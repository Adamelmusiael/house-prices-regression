```python
plt.figure(figsize=(15,8))
plt.hist(train_df['SalePrice'], bins=35, edgecolor='black')
plt.title('SalePrice distribution', size=16)
plt.xticks(ticks=np.arange(0,800000,50000))
plt.show()
```

### NOTE 
We got right-skewed distribution with mean around 180,000 and most data centered around 150,000 price and it seems like there are outliers in dataset.

```python
corr_matrix = train_df.corr()

fig, axis = plt.subplots(figsize=(15,12))
```

```python
numeric_cols = train_df.select_dtypes(include=['number'])

corr_matrix = numeric_cols.corr()

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, mask=mask, square=True)
plt.title('Train dataset correlation matrix')
```

```python
# take only the upeer triangle to avoid duplicates
upper_triangle = corr_matrix.where(
    pd.DataFrame(np.triu(np.ones(corr_matrix.shape), k=1),
                 columns=corr_matrix.columns,
                 index=corr_matrix.index).astype(bool)
)

most_corr = (upper_triangle.stack().abs().sort_values(ascending=False).head(5))
most_corr
```

### NOTE
There are some features that are impactful on SalePrice, like OverallQual, GrLivArea. Also there are few features that are highly correlated. It means that it will be worth to look into them more while testing some linear models. Since most probably for this dataset tree based models might work better I will not focus on this that much.

```python
original = train_df['SalePrice']
log = np.log(train_df['SalePrice'])
log1 = np.log(train_df['SalePrice']+1)

fig, axs = plt.subplots(2,2, figsize=(15,10))

sns.histplot(original, ax=axs[0,0])
axs[0,0].set_title('Original ')

sns.histplot(log, ax=axs[0,1])
axs[0,1].set_title('Log transformed ')

sns.histplot(log1, ax=axs[1,0])
axs[1,0].set_title('Log n+1 transformed ')

axs[1,1].set_visible(False)

plt.tight_layout()
plt.show();
```

```python
# Skewness is basically the measure the lack of the symmetry
original.skew(), log.skew(), log1.skew()
```
